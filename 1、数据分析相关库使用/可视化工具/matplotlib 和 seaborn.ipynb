## Matplotlib 可视化用法：




### 1、导入相应库

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2、Dataframe.plot() 方法：

#### 生成数据

```python
data = {'name':['congcong', 'mingming', 'meimei', 'lili', 'xingxing'],'chinese': [76, 95, 93, 92,81]}
data = pd.DataFrame(data, columns=['name','chinese'])
data['name'] = data['name'].apply(str.title)
data
```

设置支持中文显示：

```python
plt.rcParams['font.sans-serif']=['SimHei'] # 设置字体，防止中文乱码
```

```python
data.plot(kind='bar')
```

```python
data.plot(kind='bar',       # 设置你要绘制的图像类型
          legend=True,     # 是否显示图例
          figsize=(8,5),   # 设置图像大小
          x='name',        # 指定 x 轴
          y='chinese',     # 指定 y 轴
          rot = 0,         # 旋转角度
          grid=True,       # 是否显示网格
          title = '语文考试分数',   # 指定标题
          color = "red",     # 指定颜色
          alpha = 0.4,        # 设置透明度
          # label = "语文"，   # 设置图例标签
           ylim = [20, 100],    # 设置 y 轴坐标刻度
         )
```

### 3、plt.figure() 方法

#### 生成数据

```python
data = {'name':['congcong', 'mingming', 'meimei', 'lili', 'xingxing'],'chinese': [76, 95, 93, 92,81],'math': [63, 86, 92, 98, 90]}
data = pd.DataFrame(data, columns=['name','chinese','math'])
data['name'] = data['name'].apply(str.title)
data
```

```python
fig, ax = plt.subplots(2,1,figsize=(8,8))
ax[0].bar(x=data['name'],height=data['chinese'],color='m')
ax[0].set_title('语文考试分数统计')
ax[0].set_xlabel('姓名')
ax[0].set_ylabel('语文成绩')
ax[0].legend(['语文'])


ax[1].bar(x=data['name'],height=data['math'],color='c')
ax[1].set_title('数学考试分数统计')
ax[1].set_xlabel('姓名')
ax[1].set_ylabel('数学成绩')
ax[1].legend(['数学'])
plt.tight_layout()  
plt.show()
```

### 条形图

```python
fig, ax = plt.subplots(2,1,figsize=(8,8))
ax[0].barh(data['name'],data['chinese'],color='m')
ax[0].set_title('语文考试分数统计')
ax[0].set_xlabel('姓名')
ax[0].set_ylabel('语文成绩')
ax[0].legend(['语文'])


ax[1].barh(data['name'],data['math'],color='c')
ax[1].set_title('数学考试分数统计')
ax[1].set_xlabel('姓名')
ax[1].set_ylabel('数学成绩')
ax[1].legend(['数学'])
plt.tight_layout()  
plt.show()
```

### 折线图：

```python
data = {'time':['第一单元','第二单元','第三单元','期中考试','第四单元','第五单元','期末考试'],'math': [50, 63, 75, 80,88,92,88]}
data = pd.DataFrame(data, columns=['time','math'])
data
```

### 小提琴图

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## 将这些不同的集合合并到一个列表中
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# 创建 figure 实例
fig = plt.figure()

# 创建 axes 实例
ax = fig.add_axes([0,0,1,1])

# 绘图
bp = ax.violinplot(data_to_plot)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# Fixing random state for reproducibility
np.random.seed(1)


# generate some random test data
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

# plot violin plot
axes[0].violinplot(all_data,
                   showmeans=False,
                   showmedians=True)
axes[0].set_title('Violin plot')

# plot box plot
axes[1].boxplot(all_data)
axes[1].set_title('Box plot')

# adding horizontal grid lines
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data))])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')

# add x-tick labels
plt.setp(axes, xticks=[y + 1 for y in range(len(all_data))],
         xticklabels=['x1', 'x2', 'x3', 'x4'])
plt.show()
```
