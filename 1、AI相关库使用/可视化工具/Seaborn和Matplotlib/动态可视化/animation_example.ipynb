```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.animation as animation

plt.rcParams['font.sans-serif']=['SimHei']
```

```python
data = pd.read_csv('Indicators.csv')
data.head()
```

```python
data.columns
```

```python
data.drop(columns='Unnamed: 63',inplace=True)
```

```python
new_data = pd.DataFrame()
```

```python
Country_Name = ['United States','Japan','China','Australia','United Kingdom','India','Korea, Rep.']
New_Country_Name = ['United_States','Japan','China','Australia','United_Kingdom','India','Korea']
```

```python
for i in zip(Country_Name,New_Country_Name):
    print(i)
```

```python
for country_name in zip(Country_Name,New_Country_Name):
    print(country_name)
    new_data[country_name[1]] = data[data['Country Name'] == country_name[0]].iloc[:,4:].values.T.reshape(59,)
```

#### 设置变化的标准，每一次我们会重新设置图像的数据，用于图像的变化

```python
def animate(i):
    for rect, yi in zip(rects, new_data.values[i]):
        rect.set_height(yi)
    return rects

def animate2(i):
    for rect, yi in zip(rects, new_data.values[i]):
        rect.set_width(yi)
    return rects
```

#### 用于生成视频

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
```

#### 开始制作动图
    
    首先我们先定义一个基础的图，然后我们使用 animation.FuncAnimation 对我们的图进行修改，每一次修改，其实最后我们做出来的图，就是按照每一次修改根据帧数生成。

#### animation.FuncAnimation
    
*    __理解代码的关键三点：__

 *    __1、获取图像 fig 和 坐标轴对象 ax__
 
      * __fig, ax = plt.subplots()__
       
 *    __2、定义初始化函数 init 和数据更新函数 animate：__（这里没有定义）

      * 初始化函数，用于动画的初始化操作，这个可以不用为可选定义
    
      * 数据更新函数，用于告知 __animation__ 更新操作，这个函数的参数默认传入的是帧序号，但是如果 __animation.FuncAnimation__ 方法指定可选参数 __frames__ 为一个列表的话，更新函数调用时传入的值会是相应索引下的元素，此时动画终点就是取完列表元素，如果不指定 __frames__ 则会一直传入帧序号保持下去。
      
 *    __3、调用 animation.FuncAnimation 方法：__

      * 第一个参数传入图像对象，这里是 __fig__
    
      * 第二个参数默认为更新函数，这里是 __animate__
    
      * 可选参数 __init_func__ ，指定初始化函数
    
      * 可选参数 __interval__ ，指定帧间隔时间单位 ms
    
      * 可选参数 __frames__ ，指定更新序列，如果是个整数，则指定的是总的帧数
    
      * 可选参数 __save_count__ ，指定保存动画（gif或mp4）的帧数
    
      * 可选参数 __repeat__ ，指定是否循环动画
    
      * 可选参数 __blit__ ，指定是否优化绘图

```python
fig = plt.figure(figsize=(15,12))
rects = plt.bar(new_data.columns.values,new_data.values[0],color=['b','g','r','c','m','y','k'])

plt.title('各国商品贸易（占GDP的百分比%）',fontsize=26)

plt.xlabel('国家名称',fontsize=16)
plt.ylabel('商品贸易（占GDP的百分比%）',fontsize=16)

# 用于变换坐标名称
x_tick = np.linspace(0,100,6)
new_x_tick =[str(i)+'%' for i in x_tick]


plt.xticks(fontsize=14)
plt.yticks(x_tick,new_x_tick,fontsize=14)



plt.grid(ls="--")
plt.ylim(0,100)

anim = animation.FuncAnimation(fig, animate, frames=len(new_data), interval=40)
anim.save('im.mp4', writer=writer)

plt.show()
```
